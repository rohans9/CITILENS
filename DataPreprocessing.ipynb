```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# Reading the CSV file into a pandas DataFrame
crime = pd.read_csv("C:\\Users\\Rohan Sahu\\OneDrive\\Desktop\\New folder\\FIR_Details_Data.csv", low_memory=False)
```

```python
# Displaying the first few rows of the DataFrame
crime.head()
```

```python
# Displaying the column names of the DataFrame
crime.columns
```

```python
# Selecting specific columns ['Latitude', 'Longitude', 'CrimeGroup_Name'] from the DataFrame
crime = crime[['Latitude', 'Longitude', 'CrimeGroup_Name']]
crime.head()
```

```python
# Dropping rows with missing values
crime.dropna()
```

```python
# Filtering out rows where both Latitude and Longitude are zero
crime = crime.loc[crime['Latitude'] * crime['Longitude'] != 0]
```

```python
# Displaying information about the DataFrame after cleaning operations
crime.info()
```

```python
# Randomly sampling 100,000 rows from the DataFrame
crime_df = crime.sample(100000)
```

```python
# Displaying information about the sampled DataFrame
crime_df.info()
```

```python
crime_df.head()
```

```python
# Writing the sampled DataFrame to a new CSV file without including the index
crime_df.to_csv('crime_head.csv', index=False)
```
